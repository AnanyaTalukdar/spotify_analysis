```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
multi_select_cols = [
    'spotify_listening_device', 
    'music_Influencial_mood', 
    'music_lis_frequency', 
    'music_expl_method'
]

def expand_multi_select(df, column):
    expanded = df[column].str.get_dummies(sep=', ')
    expanded = expanded.add_prefix(f"{column}_")
    return expanded

expanded_df = pd.DataFrame(index=df.index)

for c in multi_select_cols:
    expanded_feature = expand_multi_select(df,c)
    expanded_df = pd.concat([expanded_df, expanded_feature], axis=1)

expanded_df.head()

```
